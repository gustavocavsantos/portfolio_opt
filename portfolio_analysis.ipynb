```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
stocks_br  = ["PETR4.SA", "VALE3.SA", "ITUB4.SA", "MGLU3.SA", "BBAS3.SA", "BBDC4.SA","B3SA3.SA", "PETR3.SA", "RENT3.SA", "ELET3.SA" ]
stocks_usa = ["META", "AAPL", "AMZN", "F", "T", "BAC", "GOOGL", "MSFT", "INTC", "CMCSA"]
stocks_eur = ["iSP.MI", "ENEL.MI", "SAN.MC", "INGA.AS", "ENI.MI", "BBVA.MC", "IBE.MC", "CS.PA", "STLA.MI","DTE.DE"]
stocks_chn = ["601899.SS","600010.SS","600795.SS", "603993.SS", "600157.SS", "601288.SS", "600050.SS", "601398.SS", "600537.SS","600777.SS"]
cmds = ['ZR', 'LE', 'KC', 'FL', 'ZC', 'TR', 'GC', 'CB', 'ZS', 'ZW']

path_list= ['results/df_actions_a2c.csv','results/df_actions_ddpg.csv','results/df_actions_ppo.csv','results/df_actions_sac.csv','results/df_actions_td3.csv']
name_list= ['A2C','DDPG','PPO','SAC','TD3']
```

```python
for i in range(0,len(path_list)):

    path = path_list[i]
    name = name_list[i]

    df = pd.read_csv(path)
    df_n = pd.DataFrame()
    df_n['Brazil'] = df[stocks_br].sum(axis=1)
    df_n['USA'] = df[stocks_usa].sum(axis=1)
    df_n['EUR'] = df[stocks_eur].sum(axis=1)
    df_n['CHINA'] = df[stocks_chn].sum(axis=1)
    df_n['COMMODITIES'] = df[cmds].sum(axis=1)
    df_n.index = df.date

    df_n.plot(figsize=(20,12), title = name)
    plt.legend(loc='upper right')
```
